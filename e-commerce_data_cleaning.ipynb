```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ecommerce_delivery_700k.csv")
df.head()
```

Check the null values

```python
df.isnull().sum()
```

Fill the null values

```python
df['ShippingCost'] = df['ShippingCost'].fillna(df['ShippingCost'].mean())
```

```python
df['CustomerRating'] = df['CustomerRating'].fillna(df['CustomerRating'].median())
```

```python
df.isnull().sum()
```

Check the Data type

```python
df.dtypes
```

Change the Data Type

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df['ShipDate'] = pd.to_datetime(df['ShipDate'])
```

```python
df['DeliveryDate'] = pd.to_datetime(df['DeliveryDate'])
```

Check the Duplicate 

```python
df.duplicated().sum()
```

Data Cleaning is Completed Successfully !

---------------------------------------------------------------------------------------------------------

**Note:**

1. I identified the missing values and imputed them using the mean and median with the `fillna()` function.
2. I checked the data types and found that the date columns were stored as strings. I converted them to the `datetime` data type.
3. I checked for duplicate records and found that there were no duplicates in the dataset.
